## Data Cleaning

### Cleaning Appointment Duration Column

```python
#Seems that mins_appointment_duration has a messy graph cuz the data is not cleaned
#Let's make a box plot in order to detect outliers

# Convert all negative values to positive values
data.loc[data['mins_appointment_duration'] < 0, 'mins_appointment_duration'] *= -1
# Modify the 'mins_appointment_duration' column in the original DataFrame
min = 5
max = 60
data.loc[~((data['mins_appointment_duration'] >= min) & (data['mins_appointment_duration'] <= max)), 'mins_appointment_duration'] = np.nan

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

box = ax.hist(data['mins_appointment_duration'].dropna())

ax.set(title="Distribution of Appointment Duration",
       ylabel="Minutes")


```

### Cleaning Practitioner Years Column

```python
# Convert all negative values to positive values
data.loc[data['practitioner_years'] < 0, 'practitioner_years'] *= -1

# Modify the 'practitioner_years' column in the original DataFrame
min_practitioner_years = 0
max_practitioner_years = 50  # Adjust the range based on your data

#Convert all outliers as null values
data.loc[~((data['practitioner_years'] >= min_practitioner_years) & (data['practitioner_years'] <= max_practitioner_years)), 'practitioner_years'] = np.nan

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Use boxplot for 'practitioner_years' after cleaning
box = ax.boxplot(data['practitioner_years'].dropna(), vert=False)

ax.set(title="Distribution of Practitioner Years",
       xlabel="Years")

# Show the plot
plt.show()
```

## Product Segment Chart

```python
# Make product segment chart
specialty_type_with_rating_grouped = data.groupby(['specialty_type','q8_rating_classification']).count()
specialty_type_with_rating_grouped.reset_index(inplace=True)

#Filter the data to only 3 specialties with high counts
filtered_data = specialty_type_with_rating_grouped[
    (specialty_type_with_rating_grouped['specialty_type'] != 'Invalid') &
    (specialty_type_with_rating_grouped['specialty_type'] != 'Supporting specialties')]
filtered_data = filtered_data.rename(columns={'q8_rating_classification': 'Overall Rating Classification'})
filtered_data['appointment_ID'] = filtered_data['appointment_ID']/data['appointment_ID'].count()*100
filtered_data['appointment_ID'] = filtered_data['appointment_ID'].apply(lambda x: math.ceil(x))
filtered_data = filtered_data.rename(columns={'appointment_ID': 'Percentage of Appointments'})
filtered_data
```

```python
# Reusable Catplot from OO Method

# Set figure
fig, ax = plt.subplots(figsize=(20, 10))  # Reducing height to avoid layout issues

# Define the color palette for low and high ratings
palette = {'Low': 'red', 'High': 'green'}

graph = sns.catplot(
    data=filtered_data,
    x='specialty_type',
    hue='Overall Rating Classification',
    y='Percentage of Appointments',
    kind='bar',
    palette=palette,
    ax=ax,  # Specify the axis
    aspect=1  # Adjust the aspect ratio as needed
)

# Adjust plot aesthetics or labels if required
plt.title('Specialty Type Distribution per Appointments',fontweight='bold')

# Adjust x-axis label font size
plt.xlabel('Specialty Type', fontsize=12)
# Set font size for x-axis tick labels ('specialty_type')
plt.xticks(fontsize=8)
# Adjust y-axis label font size
plt.ylabel('Percentage of Appointments', fontsize=12)

# Show the plot
plt.show()
```

## Quarterly Trend Chart

```python
# Convert 'appointment_date' to datetime if needed
data['appointment_date'] = pd.to_datetime(data['appointment_date'])

# Set 'appointment_date' as the DataFrame index
data.set_index('appointment_date', inplace=True)

# Group by quarters and calculate the mean 'q8_rating' and total count of 'appointment_ID'
quarterly_data = data.groupby(pd.Grouper(freq='Q')).agg({'q8_rating': 'mean', 'appointment_ID': 'count'}).reset_index()

# Rename the columns for clarity
quarterly_data.rename(columns={'q8_rating': 'mean_q8_rating', 'appointment_ID': 'total_appointment_count'}, inplace=True)

# Display the resulting DataFrame
print(quarterly_data)
```

```python
#Reusable Line Chart
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the line chart
line = ax.plot(quarterly_data.index, quarterly_data['mean_q8_rating'], marker='o', linewidth=3)

# Add a horizontal line at y=4 (green color)
ax.axhline(y=4, color='green', linestyle='--', label='Overall Rating = 4')

# Display count values as annotations with adjusted positions
for i, value in enumerate(quarterly_data['mean_q8_rating']):
    if i % 2 == 0:  # Alternate positioning for better readability
        ax.text(quarterly_data.index[i], value + 0.1, f'{value:.2f}', ha='center', va='bottom', color='black')
    else:
        ax.text(quarterly_data.index[i], value - 0.15, f'{value:.2f}', ha='center', va='top', color='black')

# Set labels and title with adjusted font properties
ax.set_xlabel('Appointment Date', fontsize=12)
ax.set_ylabel('Overall Rating Average', fontsize=12)
ax.set_title('Overall Rating Trend per Quarter', fontsize=14,fontweight='bold')

# Remove gridlines
ax.grid(False)


# Show the plot
plt.show()
```

## Correlation chart of q4-q7 ratings with q8 ratings

```python
# Make correlation chart of q4-q8 ratings
# Assign ratings to a new dataframe
scoresDataFrame = pd.DataFrame({
    'Competency_Score': data['q4_rating'],
    'Punctuality_Score': data['q5_rating'],
    'Affordability_Score': data['q6_rating'],
    'Repetition_Score': data['q7_rating'],
    'Overall_Satisfaction_Score': data['q8_rating']
})

scoresDataFrameCorr = scoresDataFrame.corr()

# Create the heatmap with the specified color range
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(scoresDataFrameCorr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, vmin=0, vmax=1, ax=ax)
ax.set_title('Correlation Of Doctor Ratings', fontweight='bold')  # Set title with bold font

# Adjust x and y tick label font sizes
ax.tick_params(axis='x', labelsize=8)
ax.tick_params(axis='y', labelsize=8)

# Show the plot
plt.show()
```

## Overlapping Histogram of Practitioner Years

```python
# Histogram of practitioner years
# Group by 'practitioner_years' and 'q8_rating_classification' and count occurrences
practitioner_years_rating_grouped = data.groupby(['practitioner_years', 'q8_rating_classification','specialty_type']).size().reset_index(name='count')

# Drop missing values (if needed)
practitioner_years_rating_grouped = practitioner_years_rating_grouped.dropna(subset=['practitioner_years', 'count','specialty_type'])

# Display the grouped data
practitioner_years_rating_grouped
```

```python
# Filter the DataFrame to include only 'Low' ratings
low_ratings_data = practitioner_years_rating_grouped[
    (practitioner_years_rating_grouped['q8_rating_classification'] == 'Low') & 
    (practitioner_years_rating_grouped['specialty_type'] == 'Surgical specialties')
]

#Reusable Histogram
# Set figure size
fig, ax = plt.subplots(figsize=(8, 10))

# Plotting the histogram for 'Low' ratings
ax.hist(low_ratings_data['practitioner_years'], bins=10, color='red', alpha=0.7, weights = low_ratings_data['count'] / low_ratings_data['count'].sum() * 100
)

# Set title and labels with adjusted font properties
ax.set_title('Distribution of Practitioner Years (Low Ratings)', fontsize=16, fontweight='bold')
ax.set_xlabel('Practitioner Years', fontsize=12)
ax.set_ylabel('Percentage of Appointments', fontsize=12)

# Set font size for tick labels
ax.tick_params(axis='both', labelsize=10)

# Exclude zero from the x-axis ticks
plt.xticks(np.arange(0, low_ratings_data['practitioner_years'].max(), step=5))  # Adjust the range based on your data

# Show the plot
plt.show()
```

```python
#Reusable Histogram
# Filter the DataFrame to include only 'High' ratings
# Filter the DataFrame to include only 'High' ratings and 'Surgical' specialty_type
high_ratings_data = practitioner_years_rating_grouped[
    (practitioner_years_rating_grouped['q8_rating_classification'] == 'High') & 
    (practitioner_years_rating_grouped['specialty_type'] == 'Surgical specialties')
]

# Set figure size
fig, ax = plt.subplots(figsize=(8, 10))

# Plotting the histogram for 'High' ratings with percentages as weights
ax.hist(high_ratings_data['practitioner_years'], bins=10, color='green', alpha=0.7, weights=high_ratings_data['count'] / high_ratings_data['count'].sum() * 100)

# Set title and labels with adjusted font properties
ax.set_title('Distribution of Practitioner Years (High Ratings)', fontsize=16, fontweight='bold')
ax.set_xlabel('Practitioner Years', fontsize=12)
ax.set_ylabel('Percentage of Appointments', fontsize=12)

# Set font size for tick labels
ax.tick_params(axis='both', labelsize=10)

# Exclude zero from the x-axis ticks
plt.xticks(np.arange(0, high_ratings_data['practitioner_years'].max(), step=5))  # Adjust the range based on your data

# Show the plot
plt.show()
```
